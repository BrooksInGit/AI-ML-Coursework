### Data Chunking

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name='cl100k_base',
    chunk_size=1000,
    chunk_overlap= 200
)
```

```python
document_chunks = pdf_loader.load_and_split(text_splitter)

len(document_chunks)

document_chunks [0].page_content

document_chunks [2].page_content

document_chunks [3].page_content
```

### Embedding

```python
embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

embedding_1 = embedding_model.embed_query(document_chunks[0].page_content)
embedding_2 = embedding_model.embed_query(document_chunks[1].page_content)

print("Dimension of the embedding vector ", len(embedding_1))
len(embedding_1)==len(embedding_2)

embedding_1,embedding_2
```

### Vector Database

```python
import os # Make sure to import os if you haven't already

out_dir = 'medical_db'
if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# This line creates and populates the vector store with your document chunks
vectorstore = Chroma.from_documents(
    documents=document_chunks,
    embedding=embedding_model,
    persist_directory=out_dir
)

# This line saves the vector store to disk
vectorstore.persist()



# This line shows the embedding function associated with the vectorstore (for inspection)
vectorstore.embeddings

# This performs a similarity search on the vector database
query = "What are the common symptoms of diabetes?"
k_value = 5

found_docs = vectorstore.similarity_search(query, k=k_value)


print(f"Found {len(found_docs)} similar documents:")
for i, doc in enumerate(found_docs):
    print(f"\n--- Document {i+1} ---")
    print(doc.page_content[:500] + "...") # Print first 500 characters of the chunk
```

### Retriever

```python
# --- Retriever setup ---
# k_value here determines how many top-k documents the retriever will fetch
# Let's use 4 as an example
retriever = vectorstore.as_retriever(
    search_type='similarity',
    search_kwargs={'k': 2}
)

# --- Getting relevant documents ---
user_query = "What are the latest treatments for type 2 diabetes?"
rel_docs = retriever.get_relevant_documents(user_query)
# rel_docs # Uncomment to inspect the retrieved documents

# You can print the content of the retrieved documents to see what the LLM will get
print(f"Retrieved {len(rel_docs)} documents relevant to the query:\n")
for i, doc in enumerate(rel_docs):
    print(f"--- Document {i+1} (Page: {doc.metadata.get('page')}) ---") # Assuming 'page' metadata exists
    print(doc.page_content[:400] + "...\n") # Print first 400 chars of content





# This is a common RAG prompt structure, combining the query with context
context = "\n".join([doc.page_content for doc in rel_docs])
prompt_template = f"""
Use the following context to answer the question. If you don't know the answer,
just say that you don't know, don't try to make up an answer.

Context:
{context}

Question: {user_query}
Answer:
"""

model_output = llm(
    prompt_template,
    max_tokens=512,
    temperature=0.1
)

# --- Displaying the LLM's output ---

# If 'llm' returns a dictionary with 'choices' as shown in your snippet, then:
print("\n--- LLM's Answer ---")
print(model_output['choices'][0]['text'])

# If 'llm' returns a simple string (common for some LangChain LLM wrappers), use:
# print(model_output)
```

### System and User Prompt Template

```python
qna_system_message = """You are a highly specialized Medical Knowledge Assistant designed to provide information exclusively from the provided Merck Manual excerpts.
Your primary role is to assist healthcare professionals by quickly accessing and summarizing relevant medical knowledge.
Strictly adhere to the information found in the 'Context' section.
If the answer or relevant information is not explicitly present in the provided context, you MUST state, "The information is not available in the provided medical manual excerpts." Do not infer, speculate, or use any outside knowledge.
Under NO circumstances should you offer medical advice, diagnoses, treatment recommendations, or personal opinions.
Maintain a neutral, objective, and factual tone at all times.
Prioritize clarity and conciseness for efficient information retrieval."""

qna_user_message_template = """
Context:
{context}

Question: {question}

Answer:
"""
```

### Response Function

```python
def generate_rag_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message  = """You are an impartial judge evaluating the 'groundedness' of an AI's answer based on a provided context.
Your task is to determine if all factual claims in the 'Answer' are directly and explicitly supported by the 'Context'.
Rate the groundedness on a scale of 0 to 1, where 1 means all claims are fully supported by the context, and 0 means there are unsupported claims or hallucinations.
Provide only the numerical score. If there are any claims in the Answer not directly found in the Context, the score should be 0.
"""
```

```python
relevance_rater_system_message = """You are an impartial judge evaluating the 'relevance' of an AI's answer to a given question, based on a provided context.
Your task is to determine if the 'Answer' directly and comprehensively addresses the 'Question' using the information available in the 'Context'.
Rate the relevance on a scale of 0 to 1, where 1 means the answer is highly relevant and directly addresses the question, and 0 means the answer is irrelevant or misses the point of the question.
Provide only the numerical score.
"""
```

```python
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}
"""
```

```python
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```
